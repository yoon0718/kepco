```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from urllib.request import urlretrieve
```

```python
driver = webdriver.Chrome() #가상의 웹드라이버
```

```python
url = "https://www.google.com/"
driver.get(url)
```

```python
driver.find_elements(By.CSS_SELECTOR, '.gb_E')[1].click() #s 를 붙여서 모두 가져와서 리스트로 나타내줌 
```

```python
elem = driver.find_element(By.NAME, 'q') #검색창 찾아보내주기
```

```python
elem.send_keys("고양이") #검색창에 입력시키기
```

```python
elem.send_keys(Keys.ENTER) #들어가기
```

```python
script_get_height = "return document.body.scrollHeight" #스크롤 맨밑으로 내리기
current_height = driver.execute_script(script_get_height)
while True:
    script_move_height = "window.scrollTo(0,document.body.scrollHeight);"
    driver.execute_script(script_move_height)
    time.sleep(1)
    height = driver.execute_script(script_get_height)
    if current_height == height:
        try:
            driver.find_element(By.CSS_SELECTOR,".LZ4I").click()
        except:
            break
    current_height = height
```

```python
elements = driver.find_elements(By.CSS_SELECTOR, '.rg_i.Q4LuWd') #사진 가져오기
i = 0
for elem in elements: #합치기
    i += 1
    try:
        elem.click()
        driver.implicitly_wait(1)
        img_url = driver.find_element(By.CSS_SELECTOR, '.sFlh5c.pT0Scc.iPVvYb').get_attribute("src")
        urlretrieve(img_url, "cat/cat" + str(i) + ".jpg")
    except:
        print(img_url)
```
