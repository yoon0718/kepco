```
import os
import glob
import unicodedata

for filename in glob.glob('*.csv'):
  nfc_filename = unicodedata.normalize('NFC', filename)
  os.rename(filename, nfc_filename) # 맥os사용자에게 파일을 받았거나 NFD방식의 이름이라고 의심된다면 NFC방식으로 바꾸어야 함.
```
