```python
def make_w(method):
  method_dict ={
      'And': [0, 0, 0, 1],
      'Nor' : [1, 0, 0, 0],
      'Or' : [0, 1, 1, 1],
      'Nand' : [1, 1, 1, 0]
  }
  data = np.array([[0,0],[1,0],[0, 1],[1, 1]])
  for w1 in np.linspace(0, 1, 20):
    for w2 in np.linspace(0, 1, 20):
      for theta in np.linspace(-1, 1, 20):
        if method in ['Nor', 'Nand']:
           if np.sum(np.where(data[:,0]*w1 + data[:,1]*w2 > theta,0,1) == method_dict.get(method)) == 4:
            return -w1, -w2, -theta
        else:
          if np.sum(np.where(data[:,0]*w1 + data[:,1]*w2 < theta,0,1) == method_dict.get(method)) == 4:
            return w1, w2, theta
```

```python
W = np.random.randn(2, 1)
b = np.zeros(1)
```

```python
def gradient(W, b):
  cols = W.shape[1]
  rows = W.shape[0]
  dW = np.zeros_like(W)
  #dW = np.zeros((W.shape[0], W.shape[1]))
  db = np.zeros_like(b)
  h = 1e-5
  for col in range(cols):
    for row in range(rows):
      y_hat = sigmoid(np.dot(X, W) + b)
      fx = loss(y_hat)
      tmp = W[row, col]
      W[row, col] += h
      y_hat = sigmoid(np.dot(X, W) + b)
      fxh = loss(y_hat)
      dv = (fxh - fx)/h
      dW[row, col] = dv
      W[row, col] = tmp
  for idx in range(len(b)):
    y_hat = sigmoid(np.dot(X, W) + b)
    fx = loss(y_hat)
    tmp = b[idx]
    b[idx] += h
    y_hat = sigmoid(np.dot(X, W) + b)
    fxh = loss(y_hat)
    db[idx] = (fxh-fx)/h
    b[idx] = tmp
  return dW, db

```

```python
W = np.random.randn(2, 1)
b = np.zeros(1)
lr = 1e-4
for _ in range(1000):
  W -=  gradient(W, b)[0]*lr
  b -= gradient(W, b)[1]*lr
W, b

```
