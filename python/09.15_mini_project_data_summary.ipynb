```python
frequency = dict()

for word in df1['context'][0].split(' '):
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
        
word = list(frequency.keys())
f = list(frequency.values())

while bool(word):
    m = min(f)
    print(word.pop(f.index(m)))
    f.remove(m)
print('')
```

```python
e = []
energy = []
nuclear = []
water = []
fire = []

for i in range(0, len(df1)):
  e.append(df1['context'][i].count('전기'))
  energy.append(df1['context'][i].count('에너지'))
  nuclear.append(df1['context'][i].count('원자력'))
  water.append(df1['context'][i].count('수력'))
  fire.append(df1['context'][i].count('화력'))

df1['elec(fre)'] = e
df1['ener(fre)'] = energy
df1['nuc(fre)'] = nuclear
df1['water(fre)'] = water
df1['fire(fre)'] = fire
df1['year'] = df1['time'].dt.year
df1['month'] = df1['time'].dt.month
df1['day'] = df1['time'].dt.day
df1['day_name'] = df1['time'].dt.day_name()
df1

df1.to_csv('윤여록_가공처리.csv', index=False)
```
