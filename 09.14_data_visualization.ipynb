```python
#Figure
#모든 그래프 구성 요소를 담고 있는 최상위 객체
#scatter() 함수로 산점도를 그릴 때 자동으로 피겨 객체가 생성
```

```python
import gdown
gdown.download('http://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
```

```python
import pandas as pd
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
#그래프 크기 바꾸기 - figure()함수를 사용하여 객체를 만들면 그래프 옵션을 조절가능
#figsize 매개변수나 dpi 매개변수를 변경하여 크기 조절
#figsize는 캔버스 크기를 dpi는 그래프자체를 키워서 본다고 생각
#그래프 기본 크기 (6, 4) = (너비, 높이) 단위-인치
```

```python
plt.figure(figsize=(9,6))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
plt.figure(dpi=144)
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
#rcParams 객체
#matplotlib 그래프의 기본값을 관리하는 객체
#객체에 담긴 값을 출력하는 것뿐만 아니라 새로운 값으로 바꿀 수 있음
```

```python
#DPI 기본값 바꾸기
plt.rcParams['figure.dpi'] = 100
```

```python
#산점도 모양 바꾸기
plt.rcParams['scatter.marker']
```

```python
#산점도 모양 바꾸기
plt.rcParams['scatter.marker'] = '*'
```

```python
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
#위의 산점도 모양을 바꾸는 방법보다 scatter()함수의 marker매개변수로 지정하는게 더 편함
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1, marker='+')
plt.show()
```
