```python
import pandas as pd
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
#서브플롯 : subplots()
#매개변수에 지정해주는 수로 서브플롯 개수 결정
fig, axs = plt.subplots(2)

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_yscale('log')

fig.show()
```

```python
#subplots함수도 figsize 매개변수를 제공
#set_title() 메서드를 이용해 그래프 제목제공 가능
fig, axs = plt.subplots(2, figsize=(6,8))

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
axs[0].set_title('scatter plot')

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_yscale('log')
axs[1].set_title('hist plot')

fig.show()
```

```python
#subplots를 가로로 나란히 표현
#subplots의 첫번째 매개변수는 행 두번째 매개변수는 열
#set_xlabel/ylabel로 축 이름도 지정
fig, axs = plt.subplots(1, 2, figsize=(10,4)) #1개의 행 2개의 열

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
axs[0].set_title('scatter plot')
axs[0].set_xlabel('number of books')
axs[0].set_ylabel('borrow count')

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_yscale('log')
axs[1].set_title('hist plot')
axs[1].set_xlabel('borrow count')
axs[1].set_ylabel('frequency')

fig.show()
```

```python
#value_counts() 메서드
#고유한 값의 등장횟수 계산
#sort_index() 메서드를 이용해 오름차순 정렬
count_by_year = ns_book7['발행년도'].value_counts()
count_by_year
```

```python
count_by_year = count_by_year.sort_index()
count_by_year
```

```python
count_by_year = count_by_year[count_by_year.index <= 2030]
count_by_year
```

```python
type(count_by_year.values)
```

```python
#주제별 도서 개수
#주제분류번호를 사용
#열에 NaN이 포함되어 있으므로 -1을 반환하는 함수를 만들어 걸러야함
#열의 값을 받아 첫 번째 문자를 반환하는 kdc_1st_char()함수를 선언 후 apply()메서드에 넣어 데이터프레임에 반복 적용
import numpy as np
def kdc_1st_char(no):
    if no is np.nan:
        return '-1'
    else:
        return no[0]
    
count_by_subject = ns_book7['주제분류번호'].apply(kdc_1st_char).value_counts()
count_by_subject
```

```python
#draw line graph
#첫번쨰 매개변수 = x축의 값
#두번째 매개변수 = y축의 값

import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100 #그래프 해상도 높이기
```

```python
plt.plot(count_by_year.index, count_by_year.values)
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
#모양과 색상 바꾸기
#linestyle매개변수 제공
#실선(-)/점선(:)/쇄선(-.)/파선(--)
#color매개변수 제공 - 색상 지정 - 16진수 컬러코드를 사용하거나 색 이름을 지정
#산점도에서 봤던 marker매개변수도 제공

plt.plot(count_by_year.index, count_by_year.values, marker='.',color='red', linestyle=':')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
#마커, 선모양, 색깔을 하나의 문자열로 합쳐서 나타낼 수 있음
plt.plot(count_by_year.index, count_by_year.values, '.r:')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year.index, count_by_year.values, '+:g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
#line graph 눈금 개수 조절 및 마커에 텍스트 표시
#앞의 그래프들은 x축 눈금이 4개만 표시
#xticks()함수를 통해 x축 눈금을 지정할 수 있음
```

```python
plt.plot(count_by_year.index, count_by_year.values, '+:g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10)) #1947~2030까지 10년씩 건너뛰면서 x축 눈금을 나타내기 위해 xticks()함수 매개변수에 파이썬 range()함수를 사용
for idx, val in count_by_year[::5].items(): #연도별 발행 도서 개수를 모두 그래프에 표시하면 너무 많기에 슬라이스 연산자(:)를 사용해 5개씩 건너뛰면서 count_by_year값을 선택
    #시리즈 객체의 items()메서드를 사용하여 인덱스와 값을 감싼 튜플을 얻는다.
    plt.annotate(val, (idx, val))
    #그래프에 값을 표시할 때는 annotate()함수를 사용 첫번째 매개변수에 그래프에 나타낼 문자열을 지정, 두번째 매개변수에 문자열이 나타날 x,y좌표를 튜플로 지정
plt.show()
```

```python
#위 그래프를 보면 마커 옆에 바로 텍스트가 있음
#텍스트 위치를 조절하기 위해 xytext매개변수를 사용
plt.plot(count_by_year.index, count_by_year.values, '+:g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10)) 
for idx, val in count_by_year[::5].items(): 
    plt.annotate(val, (idx, val), xytext=(idx+1, val+10))
    
plt.show()
```

```python

#잘 떨어져서 나타나는지 x축은 구분가능 y축은 구분 불가능(y축의 스케일이 크기 때문)
#이런 경우에는 상대적인 위치를 포인트나 픽셀단위로 지정
#xytext매개변수와 함께 textcoords 매개변수를 사용
plt.plot(count_by_year.index, count_by_year.values, '+:g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10)) 
for idx, val in count_by_year[::5].items(): 
    plt.annotate(val, (idx, val), xytext=(2,2), textcoords='offset points')
    
plt.show()
```

```python
#bar graph
#x축 값과 y축 값을 전달해주면 됨
#제목, 축 이름 등등 표시방법 동일
plt.bar(count_by_subject.index, count_by_subject.values)
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items(): 
    plt.annotate(val, (idx, val), xytext=(0,2), textcoords='offset points')
    
plt.show()
```

```python
#텍스트 정렬, 막대 조절 및 색상 바꾸기
#텍스트 위치 조절 - annotate()함수의 ha매개변수에 'center' 지정 / 기본값이 'right'이기 떄문
#텍스트 겹치는 경우가 있으므로 fontsize 매개변수로 텍스트 크기 조절
#텍스트 색깔 - color 매개변수로 조절
#막대 조절 - 두께 - width 매개변수 제공 기본값 0.8

plt.bar(count_by_subject.index, count_by_subject.values, width=0.7, color='blue')
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items(): 
    plt.annotate(val, (idx, val), xytext=(0,2), textcoords='offset points', fontsize=8, ha='center', color='green')
    
plt.show()
```

```python
#가로 막대 그래프
#barh()함수
#매개변수는 width가 아니라 height매개변수가 됨
#annotate() 함수 택스트 좌표도 x축과 y축 변경해야 함
plt.barh(count_by_subject.index, count_by_subject.values, height=0.7, color='blue')
plt.title('Books by subject')
plt.xlabel('number of books')
plt.ylabel('subject')
for idx, val in count_by_subject.items(): 
    plt.annotate(val, (val, idx), xytext=(2,0), textcoords='offset points', fontsize=8, va='center', color='green')
    #가로막대그래프에서는 va매개변수르 사용 / 기본값 baseline
plt.show()
```
