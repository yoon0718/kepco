```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd

id1 = '20'
page = '10000'
crawl_df = pd.DataFrame(columns=['big_title', 'time', 'title', 'context'])
url = f"https://www.etnews.com/news/section.html?id1={id1}"
html = urlopen(url)
soup = bs(html.read(), "html.parser")

for link in soup.select('ul > li > a'): #과학/에너지 분야 찾기
    flink_list=[]
    slink_list = [] 
    link2 = link.attrs['href']
    if "/news/section.html?id1=20&id2=" in link2:
        link3 = link2
        flink_list.append(f"https://www.etnews.com{link3}")
    for link4 in flink_list: #분야 들어가서
        url3 = f"{link4}&page=10000"
        html3 = urlopen(url3)
        soup3 = bs(html3.read(), "html.parser")
        last_page = soup3.find('a', {'href':'javascript:;'}).text #마지막 페이지 찾아주기
        for page in range(1, int(last_page)+1): # 분야별로 p.1 ~ p.last까지
            url = f"{link4}&page={page}"
            html1 = urlopen(url)
            soup1 = bs(html1.read(), "html.parser")
            for link5 in soup1.select('p > a'): #각 기사 url 찾기
                link6 = link5.attrs['href']
                if "https" not in link6:
                    slink_list.append(f"https://www.etnews.com{link6}")
                    
                        
    
    for link7 in slink_list: #기사링크 들어가기
        html2 = urlopen(link7)
        soup2 = bs(html2.read(), "html.parser")
        if link3 == '/news/section.html?id1=20&id2=020': #분야
           big_title = '과학'
        else:
            big_title = '에너지'
        time = soup2.find_all('time')[0].text[6:16] #발행날짜
        title = soup2.find_all('h2')[0].text #제목
        text_list=[]
        for k in range(len(soup2.find_all('article', {'class':'article_body_wrap'})[0].find_all('p'))):
            text_list.append(soup2.find_all('article', {'class':'article_body_wrap'})[0].find_all('p')[k].text.replace("\n", ""))
        context = ''.join(text_list)#내용
        tmp_df = pd.DataFrame([[big_title, time, title, context]], columns=crawl_df.columns)
        crawl_df = pd.concat([crawl_df, tmp_df])
        
        
        
```
