```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
apikey = 'x3r-2yN0Q8W6_tsjdOPFUQ'
tm1 = 202309100348
tm2 = pd.to_datetime(datetime.now()).strftime('%Y%m%d%H%M')


url = f"https://apihub.kma.go.kr/api/typ01/url/kma_sfctm3.php?tm1={tm1}&tm2={tm2}&stn=108&help=1&authKey={apikey}"
url
html = urlopen(url)
bs = soup(html, 'html.parser')
```

```python
print(bs)
```

```python
date_list = []
a_list = []
b_list = []

for i in (str(bs).split('\n'))[:-1]:
  if '#' not in i:
    date_list.append(i.split(' ')[0])
    a_list.append(i.split(' ')[4])
    b_list.append(i.split(' ')[6])
```

```python
date_list
```

```python
pd.DataFrame(
    {
        'date':date_list,
        '풍향':a_list,
        '풍속':b_list
    }
)
```

```python
df = pd.DataFrame(columns=['date', '풍향', '풍속'])
for i in (str(bs).split('\n'))[:-1]:
  if '#' not in i:
    date = i.split(' ')[0]
    풍향 = i.split(' ')[4]
    풍속 = i.split(' ')[6]
    tmp = pd.DataFrame([[date, 풍향, 풍속]],columns=['date', '풍향', '풍속'])
    df = pd.concat([df, tmp])
```
