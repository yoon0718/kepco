```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs

id1 = "04"
id2 = "044"
page = '1'

url = f"https://www.etnews.com/news/section.html?id1={id1}&id2={id2}&page={page}"
html = urlopen(url)
soup = bs(html.read(), "html.parser")  # html파싱
soup
```

```python
soup.select('p > a') #p태그안에 a태그 뽑아오기(href값 추출을 위함)
```

```python
soup.select('p > a')[0].attrs['href'] #잘추출되었는지 확인(0번째 p태그속 a태그의 href값)
```

```python
link_url = []  #추출한 값을 뽑아 "https"로 시작하는것이 아닌 값들을 주소에 붙여 link_url 동적배열에 저장
for links in soup.select('p > a'):
  link = links.attrs['href']
  if "https" not in link:
      link2 = link
      link_url.append(f"https://www.etnews.com{link}")
      print(link_url)
```

```python
html = urlopen(link_url[0]) #link_url에 저장된 0번째 링크에 접속해 파싱을 하고 h2태그의 텍스트를 가져온다
soup = bs(html.read(), 'html.parser')
soup.find_all('h2')[0].text
```

```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs

id1 = "04"
id2 = "044"
page = '10000'

url = f"https://www.etnews.com/news/section.html?id1={id1}&id2={id2}&page={page}" #마지막페이지 찾아주기
html = urlopen(url)
soup = bs(html.read(), "html.parser")
last_page = soup.find('a', {'href':'javascript:;'}).text
crawl_df = pd.DataFrame(columns=['title','context'])
for page in range(1, int(last_page)+1):                                           #첫페이지부터 마지막페이지까지 반복
  url = f"https://www.etnews.com/news/section.html?id1={id1}&id2={id2}&page={page}"
  html = urlopen(url)
  soup = bs(html.read(), "html.parser")
  print(f'{page} 진행중')
  link_url = []
  for links in soup.select('p > a'):
    link = links.attrs['href']
    if "https" not in link:
        link2 = link
        link_url.append(f"https://www.etnews.com{link}")

  for link in link_url:
    html = urlopen(link)
    soup = bs(html.read(), 'html.parser')
    title = soup.find_all('h2')[0].text
    context = soup.find_all('p')[0].text.replace("\n", '')
    tmp_df = pd.DataFrame([[title, context]],columns=crawl_df.columns)
    crawl_df = pd.concat([crawl_df, tmp_df])
```
